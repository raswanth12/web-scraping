```python
scoup=bs4.BeautifulSoup(response.content)
```

```python
print(scoup.prettify())
```

```python
# for extract the container form the website 
divs = scoup.find_all('div', attrs={'class': '_1AtVbE col-12-12'})

# 'class': '_2kHMtA'

 #this the class from the website coloum of the mobile 
```

```python
len(divs)
```

```python
for item in divs:
    mobileName= item.find('div',attrs={'class':'_4rR01T'})
    originalPrice= item.find('div',attrs={'class':'_3I9_wc _27UcVY'})
    offerPrice = item.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
    if mobileName is not None:
        if originalPrice is None:
            originalPrice="N/A"
        else:
            originalPrice=originalPrice
        if offerPrice is None:
            offerPrice="N/A"
        else:
            offerPrice=offerPrice
        print (mobileName.text,"\t",originalPrice.text,"\t",offerPrice.text)
```
